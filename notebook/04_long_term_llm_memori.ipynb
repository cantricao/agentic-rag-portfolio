# 🧠 Long-term Memory with Memori (Patient History Layer)

**Objective:**
In a healthcare context, remembering patient history is critical. Standard RAG only retrieves documents, but lacks "personal context".
This notebook integrates **Memori** to store and retrieve user-specific preferences and past interactions (e.g., allergies, previous diagnoses) to personalize the AI agent's responses.

**Tech Stack:**
* **Memori:** For managing user memory graph.
* **Storage:** SqlLite / ProgreSQL (managed by Memori).

```python
# Get Gemini Key
import os
from google.colab import userdata

# Retrieve the Google API key securely from Colab secrets
api_key = userdata.get('groq_api')
```

```python
from openai import OpenAI, AsyncOpenAI

# ==========================================
# 1. ENVIRONMENT SETUP & CONFIGURATION
# ==========================================

client = AsyncOpenAI(base_url="https://api.groq.com/openai/v1", api_key=api_key)
```

```python
# ==========================================
# 3. ASYNC API WRAPPER
# ==========================================

MODEL_ID = "openai/gpt-oss-20b"

import asyncio
# Define the coroutine for making API calls to GPT
async def make_api_call_to_gpt(prompt, model=MODEL_ID):
    messages = [{"role": "user", "content": prompt}]
    response =  await client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=0,
    )

    return response.choices[0].message.content
```

# No Memory

```python
task = make_api_call_to_gpt("I have a severe severe anaphylactic reaction to Penicillin.")
result = await asyncio.gather(task)
print(result)
```

```python
task = make_api_call_to_gpt("Am I allergic to anything?")
result = await asyncio.gather(task)
print(result)
```

# SQL Lite

```python
from memori import Memori
import sqlite3

def get_sqlite_connection():
    return sqlite3.connect("memori.db")


memori = Memori(conn=get_sqlite_connection).llm.register(client)
memori.attribution(entity_id="123456", process_id="test-ai-agent")
memori.config.storage.build()
```

```python
task = make_api_call_to_gpt("I have a severe severe anaphylactic reaction to Penicillin.")
result = await asyncio.gather(task)
print(result)
```

```python
task = make_api_call_to_gpt("Am I allergic to anything?")
result = await asyncio.gather(task)
result[0]
```
